```python
# h5
import h5py
f = h5py.File(filepath+'CNS005/0 SpatialReg Image Registration 0_SpatialRegistration.h5', 'r')

for k in list(f.keys()):
    dset = f[k]
    print([k,dset])
```

```python
def show_slices(slices):
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

```python
img2 = nib.load('/gpfs/data/chopralab/Glioma_segmentation/data/complete/'
                +'CNS005/102 AX MPR RECON - imageType DERIVED-SECONDARY-MPR-CSA MPR--CSAPARALLEL-M-NORM-DIS2D.nii.gz')
data = img2.get_fdata()
print(data.shape)

epi_img_data = data

slice_0 = epi_img_data[100, :, :]
slice_1 = epi_img_data[:, 200, :]
slice_2 = epi_img_data[:, :, 100]

show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Center slices for EPI image")  
plt.show()
```
